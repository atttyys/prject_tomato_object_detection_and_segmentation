```python
import optuna
from ultralytics import YOLO
import matplotlib.pyplot as plt
from optuna.visualization.matplotlib import plot_optimization_history
import pandas as pd
import numpy as np
import optuna
import tensorflow as tf
from object_detection.builders import model_builder
from object_detection.utils import config_util
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses import CategoricalCrossentropy  # ใช้ loss ที่เหมาะสมกับงานของคุณ

# กำหนดค่าคงที่และตั้งค่าเริ่มต้น
DATA_PATH = "/content/drive/MyDrive/tomato_yolov8-1/data.yaml"
MODEL_CONFIG = 'yolov5s.yaml'
MODEL_WEIGHTS = 'yolov5s.pt'
STORAGE_NAME = "sqlite:////content/drive/MyDrive/tomato_yolov8/optuna_study.db"
STUDY_NAME = "loaded_study"

optuna.logging.set_verbosity(optuna.logging.WARNING)

# สร้างฟังก์ชั่น utility เพื่อโหลดหรือสร้าง study
def get_or_create_study(study_name, storage_name):
    try:
        study = optuna.load_study(study_name=study_name, storage=storage_name)
        print("Load study successfully")
        # ส่งออก ข้อมูลจากฐานข้อมูลไว้ใน study
        return study
    except:
      # หากไม่มีจะสร้างใหม่
        return optuna.create_study(study_name=study_name, storage=storage_name, direction='minimize'),print("Wrong Load Study ,Create New Study!!")

# สร้าง objective function สำหรับการปรับแต่ง
def objective(trial):
    model = YOLO(MODEL_CONFIG)
    model = YOLO(MODEL_WEIGHTS)
    lr0 = trial.suggest_float("lr0", 1e-5, 1e-1, log=True)
    lrf = trial.suggest_float("lrf", 0.01, 1.0, log=True)
    momentum = trial.suggest_float("momentum", 0.6, 0.98)
    weight_decay = trial.suggest_float("weight_decay", 1e-5, 1e-2, log=True)
    warmup_epochs = trial.suggest_int("warmup_epochs", 0, 5.0)
    warmup_momentum = trial.suggest_float("warmup_momentum", 0.0, 0.95)
    # ทำการเทรนโมเดลด้วย hyperparameters ที่ถูกเสนอ
    results = model.train(
        data=DATA_PATH,
        epochs=50,
        imgsz=640,
        lr0=lr0,
        lrf=lrf,
        momentum=momentum,
        weight_decay=weight_decay,
        warmup_epochs=warmup_epochs,
        warmup_momentum=warmup_momentum,
        batch=24,
        device=0,
        project="/content/drive/MyDrive/tomato_yolov8",
        optimizer="SGD"
    )
    what_ = results.seg.map
    sent_ = 1 - what_
    # ใช้ค่า mAP ในการประเมินการเทรน
    return sent_

def train_with_optuna_params_yolo(study_name, storage_path, trial_id):
    # โหลด study ที่คุณสร้าง
    study = optuna.load_study(study_name=study_name, storage=storage_path)
    # รับค่าพารามิเตอร์จาก trial_id
    trial = study.trials[trial_id]
    params = trial.params
    print(params)
    lr0 = params['lr0']
    lrf = params['lrf']
    momentum = params['momentum']
    weight_decay = params['weight_decay']
    warmup_epochs = params['warmup_epochs']
    warmup_momentum = params['warmup_momentum']
    model = YOLO(MODEL_CONFIG)
    model = YOLO(MODEL_WEIGHTS)
    results = model.train(
        data=DATA_PATH,
        epochs=100,
        imgsz=640,
        lr0=lr0,
        lrf=lrf,
        momentum=momentum,
        weight_decay=weight_decay,
        warmup_epochs=warmup_epochs,
        warmup_momentum=warmup_momentum,
        batch=16,
        device=0,
        project="/content/drive/MyDrive/tomato_yolov8",
        optimizer="SGD"
    )
    # สำหรับตัวอย่างนี้ ฉันจะส่งค่าพารามิเตอร์กลับไป
    return params

def show_all_trials_in_custom_format(study_name, storage_path):
    study = optuna.load_study(study_name=study_name, storage=storage_path)
    df = study.trials_dataframe(attrs=('number', 'value', 'params', 'state'))

    # Prepare a list to hold all the rows
    data_list = []

    for index, row in df.iterrows():
        trial_id = row['number']
        objective_value = row['value']
        for col in df.columns:
            if col.startswith('params'):
                parameter_name = col.split("_", 1)[1]
                parameter_value = row[col]
                data_list.append({
                    "trial_id": trial_id,
                    "parameter_name": parameter_name,
                    "parameter_value": parameter_value,
                    "objective_value": objective_value
                })

    # Create new DataFrame using data_list
    new_df = pd.DataFrame(data_list, columns=["trial_id", "parameter_name", "parameter_value", "objective_value"])
    return new_df

# การใช้งานฟังก์ชั่น:
if __name__ == '__main__':
    # load database
    study = get_or_create_study(STUDY_NAME, STORAGE_NAME)
    # แสดงผลลัพธ์
    print(f"Best trial: {study.best_trial.params}")
    # plot กราฟแนวโน้มของค่า parameter และ ค่า แนวโน้มของ 1 - mAP
    plot_optimization_history(study)
    plt.show()


    # แสดงค่า parameter ทั้งหมด และค่า 1 - mAP หรือค่าตัวแปร ยิ่งน้อย ยิ่งดี เพราะ mAP ที่สูงจะทำให้โมเดลเรามีประสิทธิภาพ เนื่องจาก optuna จะลดค่า 1 -mAP ให้น้อยลง
    df = show_all_trials_in_custom_format(STUDY_NAME, STORAGE_NAME)
    # Save the DataFrame as a CSV
    csv_path = "/content/drive/MyDrive/tomato_yolov8/trials_data.csv"
    df.to_csv(csv_path, index=False)
    print(f"Saved DataFrame to {csv_path}")
    # ฝึกอบรมด้วยพารามิเตอร์ที่เราเลือกโดยใช้ id ของ พารามิเตอร์นั้นๆ
    # params = train_with_optuna_params(STUDY_NAME, STORAGE_NAME, 46)


    # ฝึกอบรมจากพารามิเตอร์ทั้งหมดเพื่อหาค่าที่ดีต่อไป
    # study.optimize(objective, n_trials=20)
```
